1.   Importar/Criar a tabela com o arquivo Titanic.csv
Executar consultas SQL para obter algumas informações básicas sobre os dados do Titanic:




Verificar o número total de registros (passageiros) no conjunto de dados:

```python
SELECT COUNT(*) AS total_passengers FROM passengers;
```

Verificar o número de passageiros que sobreviveram e que não sobreviveram

```python
SELECT survived, COUNT(*) AS count FROM passengers GROUP BY survived;
```

SQL02.JPG

Calcular a idade média dos passageiros

```python
SELECT AVG(age) AS average_age FROM passengers;
```

SQL03.JPG

Listar os nomes e idades dos 5 passageiros mais velhos

```python
SELECT name, age FROM passengers ORDER BY age DESC LIMIT 5;
```

SQL04.JPG

Verificar a distribuição de passageiros por classe (Pclass)

```python
SELECT pclass, COUNT(*) AS count FROM passengers GROUP BY pclass;
```

SQL05.JPG

Calcular a taxa de sobrevivência para cada classe (Pclass)

```python
SELECT pclass, SUM(survived) AS survived_count, COUNT(*) AS total_count,
       (SUM(survived) / COUNT(*)) * 100 AS survival_rate
FROM passengers
GROUP BY pclass;
```

Seguem codigos para criação de "Views".
Elas podem ser úteis para simplificar consultas frequentemente executadas e facilitar a análise de dados. Abaixo estão algumas sugestões de views que você pode criar com base nos dados do Titanic.

View para informações básicas dos passageiros

```python
CREATE VIEW vw_passenger_info AS
SELECT passenger_id, name, sex, age, embarked, pclass, survived
FROM passengers;
```

view01.JPG

View para estatísticas de sobrevivência por idade e sexo

```python
CREATE VIEW vw_survival_stats AS
SELECT sex,
       AVG(age) AS average_age,
       SUM(survived) AS survived_count,
       COUNT(*) AS total_count,
       (SUM(survived) / COUNT(*)) * 100 AS survival_rate
FROM passengers
GROUP BY sex;
```

view02.JPG

View para calcular a taxa de sobrevivência por classe e sexo

```python
CREATE VIEW vw_survival_by_class_sex AS
SELECT pclass, sex,
       SUM(survived) AS survived_count,
       COUNT(*) AS total_count,
       (SUM(survived) / COUNT(*)) * 100 AS survival_rate
FROM passengers
GROUP BY pclass, sex;
```
