#Em metodologia cientifica, qual a diferena dos grupos de Piloto e Controle?
D um exemplo.

Grupo Controle:

Um grupo no qual os participantes no recebem o tratamento experimental. Ele serve como uma referncia para comparao com o grupo que recebe o tratamento. O grupo controle  projetado para isolar e avaliar o efeito do tratamento, permitindo que os pesquisadores determinem se as mudanas observadas so atribuveis ao tratamento em si ou a outros fatores.

Neste estudo, o grupo controle seria composto por indivduos que recebem um placebo (uma substncia sem efeito teraputico) em vez do novo medicamento. Eles fornecem uma linha de base para comparar com o grupo experimental que recebe o medicamento real.


---

Grupo Piloto:

Refere-se a uma fase inicial ou prvia de um estudo experimental. O grupo piloto pode ser usado para testar a viabilidade do protocolo do estudo, identificar problemas potenciais e realizar ajustes antes da implementao completa do experimento. Ele no faz parte da anlise final dos resultados e no est includo nas concluses finais do estudo.

Antes de realizar o estudo completo, o pesquisador pode realizar um grupo piloto com um pequeno nmero de participantes para testar o protocolo do estudo, identificar possveis desafios logsticos ou ajustar a dosagem do medicamento, se necessrio. Os resultados do grupo piloto no seriam includos na anlise principal do estudo.

#9) 9)	Calcule o KS da tabela abaixo:

image.png

```python
# Dados fornecidos
scores = [1, 2, 3, 4, 5]
nao_pagos = [880, 750, 670, 640, 600]
pagamentos = [100, 250, 430, 510, 600]

# Adicionando o total
total_nao_pagos = sum(nao_pagos)
total_pagamentos = sum(pagamentos)

# Criar um DataFrame
data = {'Score': scores, 'Qtde No Pagos': nao_pagos, 'Qtde Pagamentos': pagamentos}
df = pd.DataFrame(data)

# Adicionar a linha de total
df.loc['Total'] = ['-', total_nao_pagos, total_pagamentos]

# Exibir o DataFrame
print(df)

```

#KS

```python
# Calcular as distribuies acumuladas (CDF)
df['Cumulative No Pagos'] = df['Qtde No Pagos'].cumsum() / df.loc['Total', 'Qtde No Pagos']
df['Cumulative Pagamentos'] = df['Qtde Pagamentos'].cumsum() / df.loc['Total', 'Qtde Pagamentos']

# Calcular o KS
df['KS'] = abs(df['Cumulative No Pagos'] - df['Cumulative Pagamentos'])

# Encontrar o ponto de maior diferena (KS mximo)
ks_max_index = df['KS'].idxmax()

# Exibir o DataFrame com o resultado do KS
print(df)
print("\nKS Mximo:", df.loc[ks_max_index, 'KS'])
```

11)	A partir das tabelas de banco de dados acima, construa uma s query (T-SQL) como se fosse preencher quadro abaixo (de exemplo), seguindo as regras:

	Somente o produto Carto PL;

	Pagamentos (se houver) na safra;

	Atraso de 10 a 60 dias.


image.png

SQL

```python
declare @busca_produto varchar(50) = 'cartao pl'

select score,
		cpf,
		(divida - valor_pago) as divida,
		case when qtde_pagto = 0 then 0 else valor_pago / qtde_pagto end as ticket_medio,
		qtde_pagto,
		valor_pago
from	(
		select t1.score,
				t1.cpf,
				t1.divida,
				(select count(1) from tb_pagamento x where x.cpf = t1.cpf) as qtde_pagto,
				(select ISNULL(sum(valor_pago),0) from tb_pagamento x where x.cpf = t1.cpf) as valor_pago
		from tb_mailing t1
		inner	join tb_produto t2
		on	t2.cpf = t1.cpf
		where t2.produto = @busca_produto
		) as tt
order by score
```
